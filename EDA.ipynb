## 1. Exploratory Data Analysis

### i. Checking Whether in missing value columns we have more missing values or no missing values:

```python
## Relation with target variable(SalePrice)
for feature in na_features:
    data = df_train.copy()
    # let's make a variable that indicates 1 if the observation was missing or zero otherwise
    data[feature] = np.where(data[feature].isnull(),1,0)
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

- #### So we can see that missing values have great relationship with target variable as for missing values the price are also high. So we have to fill the na values with some meaningful values.

### iii. Year/Datetime Features

```python
year_features = [feature for feature in df_train.columns if 'Yr' in feature or 'Year' in feature]
year_features
```

```python
for x in year_features:
    print(x, df_train[x].unique())
```

### We will check whether there is a relation between year the house is sold and the sales price.

```python
plt.title('House Price vs Year of Sold')
df_train.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year of Sold')
plt.ylabel('Median Price')
plt.show()
```

### Here we will compare the difference between All years feature with SalePrice

```python
for feature in year_features:
    if feature != 'YrSold':
        data = df_train.copy()
        #We will calculate the difference b/w year sold and year variables
        data[feature]  = data['YrSold']-data[feature]
        plt.scatter(data[feature],data["SalePrice"])
        plt.title(feature)
        plt.show()
```

### Numerical variables are usually of 2 type
- ### Continous variable and Discrete Variables

### iv. Discrete Features

```python
## Find discrete variables
discrete_feature = [feature for feature in numerical_features if len(df_train[feature].unique()) < 25
                    and  feature not in year_features+['Id']]
discrete_feature
```

```python
print("There are",len(discrete_feature), "discrete features.")
```

```python
df_train[discrete_feature].head()
```

### Lets Find the realtionship between discrete features and Sale Price

```python
for feature in discrete_feature:
    data = df_train.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

### v. Continuous Variable

```python
continuous_features = [feature for feature in numerical_features if feature not in year_features+discrete_feature+['Id']]
continuous_features
```

```python
print("There are",len(continuous_features), "continuous features.")
```

### Lets Find the realtionship between continuous features and Sale Price

```python
for feature in continuous_features:
    data = df_train.copy()
    data[feature].hist(bins=20)
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.title(feature)
    plt.show()
```

### vi. Data Normalization

```python
for feature in continuous_features:
    data = df_train.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature] = np.log(data[feature])
        data['SalePrice'] = np.log(data['SalePrice'])
        plt.scatter(data[feature],data['SalePrice'])
        plt.title(feature)
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

### vii. Checking Outliers

```python
for feature in continuous_features:
    data = df_train.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        sns.boxplot(y = feature,data=data)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
    
```

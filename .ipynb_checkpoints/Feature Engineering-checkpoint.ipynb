### Let's Merge them so that while cleaning train data we can also clean test data

```python
df = pd.concat([df_train,df_test],axis=0)
df.head()
```

```python
print(df.shape)
```

## 1. Feature Engineering

### i. Missing Values

```python
df.isnull().sum()
```

```python
na_features = [features for features in df.columns if df[features].isnull().sum()>=1]
na_features = na_features[:-1] # To skip our target
na_features
```

```python
for feature in na_features:
    print(f"{feature} has {np.round(df[feature].isnull().mean(),4)} % of missing values.")
```

### ii. Replace missing values of categorical features

```python
categorical_feature_nan = [feature for feature in df.columns if df[feature].isnull().sum()>=1 and df[feature].dtypes=='O']
for feature in categorical_feature_nan:
    print("{}: {}% missing value".format(feature,np.around(df[feature].isnull().mean(),4)))
```

```python
def replace_null(dataset,cat_features):
    data = dataset.copy()
    data[cat_features] = data[cat_features].fillna("Missing")
    return data
df = replace_null(df,categorical_feature_nan)
```

### iii. Replace Values for numerical features

```python
numerical_with_nan=[feature for feature in df.columns if df[feature].isnull().sum()>=1 and df[feature].dtypes!='O']

## We will print the numerical nan variables and percentage of missing values

for feature in numerical_with_nan:
    print("{}: {}% missing value".format(feature,np.around(df[feature].isnull().mean(),4)))
```

```python
## SKip SalePrice column
numerical_with_nan = numerical_with_nan[:-1]
numerical_with_nan
```

```python
for feature in numerical_with_nan:
    df[feature+'nan'] = np.where(df[feature].isnull(),1,0)
    df[feature] = df[feature].fillna(df[feature].median)
df.isnull().sum()
```
